### What are the ten most common movie names of all time?
#### Ans
- We need to count the repeatation of name in coloumn title. and can print top 10 frequent entries.

### Which three years of the 1930s saw the most films released?

- We need to count the repeatation of name in coloumn title. and can print top 3 using head(3).
- For year calculation, 1930/10 gives 193, So when we use floor division to years with 10 and if the  and out put is equals to 193 will give us range from 1930 to 1939

```python
common_movie_30 = title_data[title_data.year // 10 == 193]
print(common_movie_30.year.value_counts().head(3))
```

### Plot the number of films that have been released each decade over the history of cinema.

- For Range of year calculation, XXYY/10 gives XXY, afetr multiply it gives XXY0.
- Bar Chart used for study of Values count (Give frequency or repetation values)

```python
y1=(title_data.year // 10 * 10).value_counts()
plt.ylabel('number of films')
plt.xlabel('Years')
plt.title('number of films per decade')
y1.sort_index().plot(kind='barh')
```

### Plot the number of "Hamlet" films made each decade.

```python
Hamlet_F= title_data[title_data.title == "Hamlet"]
plt.ylabel('Hamlet films')
plt.xlabel('Years')
plt.title('number of Hamlet films per decade')
(Hamlet_F.year // 10*10).value_counts().sort_index().plot(kind='barh')
```

### Plot the number of "Rustler" characters in each decade of the history of film.

### Plot the n-values of the roles that Judi Dench has played over her career.

```python
plt.figure(figsize=(10,5))
J = title_data[title_data.name == 'Judi Dench']
x=J.year
y=J.n
plt.ylabel('n-Value')
plt.xlabel('Years')
plt.title('number of Play by Judi Dench')
plt.plot(x, y, 'c *',ms=10)
```

### Plot the n-values of Cary Grant's roles through his career.

```python
cg = title_data[title_data.name == 'Cary Grant'].sort_values('year')
cg = cg[cg.n.notnull()]
cg.plot(x='year', y='n', kind='scatter')
plt.ylabel('n-Value')
plt.xlabel('Years')
plt.title("number of Play by Cary Grant's")
```

### Plot the n-value of the roles that Sidney Poitier has acted over the years.

```python
sp = title_data[title_data.name == 'Sidney Poitier'].sort_values('year')
sp = sp[sp.n.notnull()]
sp.plot(x='year', y='n', kind='scatter')
plt.ylabel('n-Value')
plt.xlabel('Years')
plt.title("number of Play by Sidney Poitier")
```
